## Calculate the theoretical steady-state Average Run Lengths (ARL1) to compare with the observed average detection delays from the experiments

##### The theoretical steady state ARL computations for the two-sided CUSUM charts from [1], implemented in the statistical process control (SPC) package for the R software was used to compare with the observed ADD in our experiments.

Refer:
[1] Crosier, R.B., 1986. A new two-sided cumulative sum quality control scheme. Technometrics, 28(3), pp.187-194.

```python
import rpy2
#from rpy2.robjects.packages import importr
#import rpy2.robjects.packages as rpackages
# import R's "base" package
#base = importr('base')
# import R's "utils" package
#utils = importr('utils')
# select a mirror for R packages
#utils.chooseCRANmirror(ind=1) # select the first mirror in the list

# import rpy2's package module
import rpy2.robjects.packages as rpackages

# import R's utility package
utils = rpackages.importr('utils')
spc = rpackages.importr('spc')
# select a mirror for R packages
utils.chooseCRANmirror(ind=1) # select the first mirror in the list

# R package names
packnames = ('ggplot2', 'hexbin', 'lazyeval', 'cusumcharter', 'RcppCNPy', 'spc')

# R vector of strings
from rpy2.robjects.vectors import StrVector

# Selectively install what needs to be install.
# We are fancy, just because we can.
names_to_install = [x for x in packnames if not rpackages.isinstalled(x)]
if len(names_to_install) > 0:
    utils.install_packages(StrVector(names_to_install))

import rpy2.robjects.lib.ggplot2 as gp
#import rpy2.robjects.lib.spc as spc
import pandas as pd
import numpy as np
```

```python
spc = rpackages.importr('spc')

h = 4 #Default choice for threshold
k = 0.5 #Default choice 
mu_1 = 0.66
ARL_1_h4k05 = spc.xcusum_ad_(k=0.5,h=4,mu1=mu_1,mu0=0,sided="two")
ARL_1_h5k05 = spc.xcusum_ad_(k=0.5,h=5,mu1=mu_1,mu0=0,sided="two")
ARL_1_h4k1  = spc.xcusum_ad_(k=1,  h=4,mu1=mu_1,mu0=0,sided="two")
ARL_1_h5k1  = spc.xcusum_ad_(k=1,  h=5,mu1=mu_1,mu0=0,sided="two")
ARL_1_h5k1
```

```python
#Print a summary table of reference values for an intended ARL0 with normalized threshold, h = 4
print("Theoretical ARL_1:")
summary_table        = pd.DataFrame(columns=["h", "k", "ARL_1"])
summary_table.loc[1] = [4,  0.5, np.round(ARL_1_h4k05[0],decimals=4)] 
summary_table.loc[2] = [5,  0.5, np.round(ARL_1_h5k05[0],decimals=4)]
summary_table.loc[3] = [4,  1,   np.round(ARL_1_h4k1[0], decimals=4)]
summary_table.loc[4] = [5,  1,   np.round(ARL_1_h5k1[0], decimals=4)]

#Show the summary table
summary_table
```
